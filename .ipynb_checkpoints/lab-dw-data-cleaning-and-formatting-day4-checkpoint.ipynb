## Exercise 2: Cleaning invalid Values

The dataset contains columns with inconsistent and incorrect values that could affect the accuracy of our analysis. Therefore, we need to clean these columns to ensure that they only contain valid data.

Note that this exercise will focus only on cleaning inconsistent values and will not involve handling null values (NaN or None).

*Hint*:
- *Gender column contains various inconsistent values such as "F", "M", "Femal", "Male", "female", which need to be standardized, for example, to "M" and "F".*
- *State abbreviations be can replaced with its full name, for example "AZ": "Arizona", "Cali": "California", "WA": "Washington"*
- *In education, "Bachelors" could be replaced by "Bachelor"*
- *In Customer Lifetime Value, delete the `%` character*
- *In vehicle class, "Sports Car", "Luxury SUV" and "Luxury Car" could be replaced by "Luxury"*

```python
#Gender column contains various inconsistent values such as "F", "M", "Femal", "Male", "female", which need to be standardized, for example, to "M" and "F".
data['gender'].unique()
```

```python
def cleangender(x):
    if x in ['M', 'MALE']:
        return 'M'
    elif x in ['F', 'Femal', 'female']:
        return 'F'
    else:
        return 'U'
data['gender'] = list(map(cleangender, data['gender']))
data['gender'].unique()
data
```

```python
#State abbreviations be can replaced with its full name, for example "AZ": "Arizona", "Cali": "California", "WA": "Washington"
data['state'].unique()
```

```python
def changest(y):
    if y in ['Cali', 'California']:
        return 'California'
    elif y in ['AZ', 'Arizona']:
        return 'Arizona'
    elif y in ['WA', 'Washington']:
        return 'Washington'
    elif y in ['Nevada']:
        return 'Nevada'
    elif y in ['Oregon']:
        return 'Oregon'
    else:
        return 'Unknown'
    
data['state'] = list(map(changest, data['state']))
data['state'].unique()
```

```python
data
```

```python
#In education, "Bachelors" could be replaced by "Bachelor"
data['education'] = data['education'].replace('Bachelors', 'Bachelor')
```

```python
data['education'].unique()
```

```python
#In Customer Lifetime Value, delete the % character
data['customer_lifetime_value'] = data['customer_lifetime_value'].str.replace('%', ' ')
```

```python
data
```

```python
#in vehicle class, "Sports Car", "Luxury SUV" and "Luxury Car" could be replaced by "Luxury"
data['vehicle_class'].unique()
```

```python
Luxury = {'Luxury SUV': 'Luxury', 'Sports Car':'Luxury', 'Luxury Car': 'Luxury'}
```

```python
data['vehicle_class'] = data['vehicle_class'].replace(Luxury)
```

```python
data['vehicle_class'].unique()
```

```python
data['vehicle_class'].value_counts(dropna=False)
```

**Lab | Customer Analysis Round 5**

Processing Data
(Further processing...)

X-y split. Remember that at this time yoou need the complete dataset, both numerical and categorical in the X features.
X split into X_numerical and X_categorical
Scaling (X_numerical).

```python
from sklearn.preprocessing import MinMaxScaler 
from sklearn.preprocessing import StandardScaler
y= data_cln['total_claim_amount']
X= data_cln.drop(['total_claim_amount'], axis=1)
```

```python
X_numerical = X.select_dtypes(np.number)
X_categorical = X.select_dtypes(object)
```

```python
transformer = MinMaxScaler().fit(X_numerical)
x_norm = transformer.transform(X_numerical)
print(x_norm.shape)
mindata= pd.DataFrame(x_norm, columns=X_numerical.columns)
mindata.head()
```

**Round 6**

One Hot/Label Encoding (categorical).

Concat DataFrames

```python
X_categorical.head(10)
```

```python
X_categorical1 = X_categorical.drop('customer', axis=1)
```

```python

from sklearn.preprocessing import OneHotEncoder

encoder = OneHotEncoder(drop='first').fit(X_categorical1)  
encoded = encoder.transform(X_categorical1).toarray() 
encoded_columns = OneHotEncoder(drop='first').fit(X_categorical1).get_feature_names_out(input_features=X_categorical1.columns) 

onehot_encoded_X = pd.DataFrame(encoded, columns = encoded_columns) 
onehot_encoded_X.head()
```

```python
conc_data= pd.concat([onehot_encoded_X, mindata ], axis=1)
conc_data
```

```python
#Train Test Split

from sklearn.model_selection import train_test_split as tts

X_train, X_test, y_train, y_test=tts(conc_data, y, test_size=.2)
```

```python
X_train
```

```python
y_train
```

```python
#Train and run model
from sklearn.linear_model import LinearRegression as linreg

lm = linreg()   
lm.fit(X_train, y_train)  
y_pred = lm.predict(X_test)  
```

```python
print("b0 = " , lm.intercept_)
print("b1 = " , lm.coef_)
```

```python
#R2 Model Validation
print ('train R2: {} -- test R2: {}'.format(lm.score(X_train, y_train),lm.score(X_test, y_test)))
```

```python
#Mean Squared Error
from sklearn.metrics import mean_squared_error as mse


train_mse=mse(lm.predict(X_train), y_train)
test_mse=mse(y_pred, y_test)

print ('train MSE: {} -- test MSE: {}'.format(train_mse, test_mse))
```

```python
#Root Mean Squared Error
print ('train RMSE: {} -- test RMSE: {}'.format(train_mse**.5, test_mse**.5))
```

```python
#Mean Absolute Error
from sklearn.metrics import mean_absolute_error as mae

train_mae=mae(lm.predict(X_train), y_train)
test_mae=mae(y_pred, y_test)

print ('train MAE: {} -- test MAE: {}'.format(train_mae, test_mae))
```
